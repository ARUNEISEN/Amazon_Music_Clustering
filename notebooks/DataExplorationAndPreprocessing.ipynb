```python
## Create a function to extract year alone in the release data feature

def extract_year(date_value):
    try:
        return pd.to_datetime(date_value, errors='coerce').year
    except:
        return int(str(date_value[:4]))
```

```python
df['release_year']= df['release_date'].apply(extract_year)
```

```python
df.drop(['release_date'], axis=1, inplace=True)
```

```python
## Apply Label encoding for id_artists and genres columns
from sklearn.preprocessing import LabelEncoder
label_encoder_id_artists = LabelEncoder()
df['id_artists'] = label_encoder_id_artists.fit_transform(df['id_artists'])
label_encoder_genres = LabelEncoder()
df['genres'] = label_encoder_genres.fit_transform(df['genres'])
df       
```

```python
df.info()
```

```python
numeric_cols = ['duration_ms','loudness','tempo','valence','acousticness','instrumentalness','speechiness','liveness']

for col in numeric_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    df_clean = df_clean[(df_clean[col] >= lower_limit) & (df_clean[col] <= upper_limit)]
```

## Feature Section
#### Selected features that represent the sound of a song and are useful for clustering.


```python
# Selected sound-related features
sound_features = [
    'danceability', 'energy', 'loudness', 'speechiness',
    'acousticness', 'instrumentalness', 'liveness',
    'valence', 'tempo', 'key', 'mode', 'time_signature', 'duration_ms'
]

df_clean = df_clean[sound_features]
df_clean
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_clean_scaled = scaler.fit_transform(df_clean)
df_clean_scaled = pd.DataFrame(df_clean_scaled, columns=df_clean.columns)
df_clean_scaled
```

```python
with open("scaler.pkl", "wb") as file:
    pickle.dump(scaler, file)
```

```python
# Save cleaned DataFrame as pickle file
with open('Cleaned_Sclaed_AmazonMusic_data.pkl', 'wb') as f:
    pickle.dump(df_clean_scaled, f)
```

## Outliers Removal & Feature Selection - Project Recommendation



```python
dataset = df.copy()
dataset.head()
```

```python
numeric_cols = ['duration_ms','loudness','tempo','valence','acousticness','instrumentalness','speechiness','liveness']

for col in numeric_cols:
    Q1 = dataset[col].quantile(0.25)
    Q3 = dataset[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    dataset = dataset[(dataset[col] >= lower_limit) & (dataset[col] <= upper_limit)]
```

```python
dataset.shape
```

```python
# Selected sound-related features
sound_features = [
    'danceability', 'energy', 'loudness', 'speechiness',
    'acousticness', 'instrumentalness', 'liveness',
    'valence', 'tempo', 'duration_ms'
]

dataset = dataset[sound_features]
dataset
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
dataset_scaled = std_scaler.fit_transform(dataset)
dataset_scaled = pd.DataFrame(dataset_scaled, columns=dataset.columns)
dataset_scaled
```

```python
with open("std_scaler.pkl", "wb") as file:
    pickle.dump(std_scaler, file)

# Save cleaned DataFrame as pickle file
with open('AmazonMusic_dataset_scaled.pkl', 'wb') as f:
    pickle.dump(dataset_scaled, f)
```
